## Gradient Boosting Classifier

```python
# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=123)
```

```python
# Display target balance in the training dataset
print(collections.Counter(y_train))
fig, ax = plt.subplots(figsize=(6,4))
sns.countplot(x="y", data=pd.DataFrame(data={'y':y_train}), ax=ax)
```

```python
# Applyting SMOTE to generare new instances (oversampling)
sm = SMOTE(random_state=123, sampling_strategy=1.0)
X_train2, y_train2 = sm.fit_resample(X_train, y_train)

# Display target balance in the training dataset
print(collections.Counter(y_train2))
fig, ax = plt.subplots(figsize=(6,4))
sns.countplot(x="y", data=pd.DataFrame(data={'y':y_train2}), ax=ax)
```

### Initial Model Training 

####  Train Initial Model

```python
# Let's define the initial model
initial_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.5, max_depth=3, random_state=123)
```

#### Cross-Validation

```python
# We applied Cross-validation to help us understanding how well our initial model is performing before any tuning

cv_scores = cross_val_score(initial_model, X_train2, y_train2, cv=5, scoring='roc_auc')
print(f"Cross-Validated AUC Scores: {cv_scores}")
print(f"Mean AUC: {cv_scores.mean():.3f}")
```

**Conclusion**
- We decided to use cross-validation to ensure that the performance metric is reliable, ensuring model robusteness.
- AUC represents the area under the ROC curve, so the higher value, the better.
- Each value in the 'cross-Validated AUC Scores' represents the AUC score for one fold of the cross-validation ( in this case are 5 because cv=5)
- A mean AUC of 0.96 indicates that our initial model can distinguish very well between responders and non-responders, but we still have to be carefull with the overfitting issue.

### Hyperparameter Tuning

####  Grid Search with Cross-Validation

```python
# Let's performe hyperparameter tuning

param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1],
    'max_depth': [3, 4, 5]
}

grid_search = GridSearchCV(estimator=GradientBoostingClassifier(random_state=123), param_grid=param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train2, y_train2)

print("Best parameters:", grid_search.best_params_)
print("Best cross-validated AUC score: ", grid_search.best_score_)
```

**Conclusion:**
- By performing grid search, we can conclude what is the best combination of hyperparameters that maximize the model's performance on the training data, validated through cross-validation.
- We can see that the best combination of parameters found by the grid search are 'learning_rate': 0.75, 'max_depth': 5, 'n_estimators': 200

### Final Model Training

```python
best_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.075, max_depth=3, random_state=123)
best_model.fit(X_train2, y_train2)
```

#### Threshold Tuning

```python
# Let's find the optimal threshold that maximizes the F1 Score

precision, recall, thresholds = precision_recall_curve(y_test, y_pred_test_proba)

f1_scores = 2 * (precision * recall) / (precision + recall)

optimal_threshold = thresholds[np.argmax(f1_scores)]

print(f"Optimal Threshold: {optimal_threshold}")
```

**Conclusion**
- By applying the optimal threshold (0.12), we can ensure that our model uses the threshold that better balances precision and recall, leading to improved performance based on our specific goals that are increase responses (recall) but also reduce expenses (precision)

```python
# Evaluate the model with the new threshold


def performanceMetricsDF(metricsObj,yTrain, yPredTrain, yTest, yPredTest):
    measures_list = ['ACCURACY', 'PRECISION', 'RECALL', 'F1 SCORE', 'AUC']
    train_results = [accuracy_score(yTrain, yPredTrain),
                     precision_score(yTrain, yPredTrain),
                     recall_score(yTrain, yPredTrain),
                     f1_score(yTrain, yPredTrain),
                     roc_auc_score(yTrain, yPredTrain)
                    ]
    test_results = [accuracy_score(yTest, yPredTest),
                    precision_score(yTest, yPredTest),
                    recall_score(yTest, yPredTest),
                    f1_score(yTest, yPredTest),
                    roc_auc_score(yTest, yPredTest)
                   ]
    resultsDF = pd.DataFrame({'Measure': measures_list, 'Train': train_results, 'Test': test_results})
    print(resultsDF)
    return resultsDF
```

```python
# Confusion matrix plotting function

def make_confusion_matrix(cf, group_names=None, categories='auto', count=True, percent=True, cbar=True,
                          xyticks=True, xyplotlabels=True, sum_stats=True, figsize=None, cmap='Blues', title=None):
    blanks = ['' for i in range(cf.size)]
    if group_names and len(group_names) == cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks
    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks
    if percent:
        group_percentages = ["{0:.2%}".format(value) for value in cf.flatten() / np.sum(cf)]
    else:
        group_percentages = blanks
    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip(group_labels, group_counts, group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0], cf.shape[1])
    if sum_stats:
        accuracy = np.trace(cf) / float(np.sum(cf))
        if len(cf) == 2:
            precision = cf[1, 1] / sum(cf[:, 1])
            recall = cf[1, 1] / sum(cf[1, :])
            f1_score = 2 * precision * recall / (precision + recall)
            stats_text = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                accuracy, precision, recall, f1_score)
        else:
            stats_text = "\n\nAccuracy={:0.3f}".format(accuracy)
    else:
        stats_text = ""
    if figsize is None:
        figsize = plt.rcParams.get('figure.figsize')
    if not xyticks:
        categories = False
    plt.figure(figsize=figsize)
    ax = sns.heatmap(cf, annot=box_labels, fmt="", cmap=cmap, cbar=cbar, xticklabels=categories, yticklabels=categories)
    if xyplotlabels:
        plt.ylabel('True label')
        plt.xlabel('Predicted label' + stats_text)
    else:
        plt.xlabel(stats_text)
    if title:
        plt.title(title)

# Show the confusion matrix
cf = confusion_matrix(y_test, y_pred_test_optimal)
labels = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
categories = ['0', '1']
make_confusion_matrix(cf, group_names=labels, categories=categories, cmap='Blues')
```
